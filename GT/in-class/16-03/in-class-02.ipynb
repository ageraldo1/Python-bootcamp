```python
import pandas as pd
```

### 1 Read in both of the CSV files and print out their DataFrames.

```python
bitcoin = pd.read_csv('resources/bitcoin_cash_price.csv')
dash_price = pd.read_csv('resources/dash_price.csv')
```

```python
bitcoin.head(5)
```

```python
dash_price.head(5)
```

### 2. Perform an inner merge that combines both DataFrames on the "Date" column.

```python
combined = pd.merge(bitcoin, dash_price, how='inner', on='Date')

combined.head(5)
```

### 3.Rename the columns within the newly merged DataFrame so that the headers are more descriptive.

```python
new_columns = {"Open_x" : "Bitcoin_Open", 
               "High_x" : "Bitcoin_High",
               "Low_x"  : "Bitcoin_Low",
               "Close_x" : "Bitcoin_Close",
               "Volume_x" : "Bitcoin_Volume",
               "Market Cap_x" : "Bitcoin_Market_Cap",
               "Open_y": "Dash_Open",
               "High_y" : "Dash_High",
               "Low_y" : "Dash_Low",
               "Close_y" : "Dash_Close",
               "Volume_y" : "Dash_Volume",
               "Market Cap_y" : "Dash_Market_Cap"
              }

combined.rename(columns = new_columns, inplace=True)

combined.head(5)

# alternative : use suffixes option during the merge
```

### 4. Create a summary table that includes the following information: Best Bitcoin Open, Best Dash Open, Best Bitcoin Close, Best Dash Close, Total Bitcoin Volume, Total Dash Volume.

```python
best_bc_open  = combined['Bitcoin_Open'].min()
best_bc_close = combined['Bitcoin_Close'].max()
tot_bc_volume = combined['Bitcoin_Volume'].sum() / 1000000

best_dash_open = combined['Dash_Open'].min()
best_dash_close = combined['Dash_Close'].max()
tot_dash_volume = combined['Dash_Volume'].sum() / 1000000


dict_summary = {
    "Best Bitcoin Open": [best_bc_open],
    "Best Dash Open": [best_dash_open],
    "Best Bitcoin Close" : [best_bc_close],
    "Best Dash Close" :  [best_dash_close],
    "Total Bitcoin Volume" : [tot_bc_volume],
    "Total Dash Volume" : [tot_dash_volume]
}

summary = pd.DataFrame(dict_summary)

summary.head(5)
```
