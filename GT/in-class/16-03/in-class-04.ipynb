### 1.Read in the CSV file provided and print it to the screen.

```python
import pandas as pd
```

```python
df = pd.read_csv('resources/ted_talks.csv')

df.head(5)
```

### 2.Find the minimum "views" and maximum "views".

```python
min_views = df['views'].min()
max_views = df['views'].max()

print (f'Min : {min_views} - Max : {max_views}')
```

### 3.Using the minimum and maximum "views" as a reference, create 10 bins in which to slice the data.


```python
step = int((max_views - min_views) / 10)

pieces = list(range(min_views, max_views, step ))

pieces[len(pieces) -1 ] = max_views

pieces
```

```python
groups = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
```

### 4. Create a new column called "View Group" and fill it with the values collected through your slicing.

```python
df['View Group'] = pd.cut(df['views'], pieces, labels=groups)

df.head(10)
```

### 5. Group the DataFrame based upon the values within "View Group".

```python
df_grp = df.groupby(by='View Group')
```

### 6. Find out how many rows fall into each group before finding the averages for "comments", "duration", and "languages".

```python
df_grp.count()
```
