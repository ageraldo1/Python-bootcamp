```python
import pandas as pd
```

### 1. Read our Kickstarter data into pandas 

```python
df = pd.read_csv('resources/KickstarterData.csv')

df.head(5)
```

### 2. Get a list of all of our columns for easy reference

```python
df.columns
```

### 3.  Extract "name", "goal", "pledged", "state", "country", "staff_pick", "backers_count", and "spotlight"

```python
c_list = ['name', 'goal', 'pledged', 'state', 'country', 'staff_pick', 'backers_count', 'spotlight']

df_ks = df[c_list]

df_ks.head(10)
```

### 4. Remove projects that made no money at all

```python
df_ks[df_ks['pledged'] == 0].count()
```

```python
cond1 = df_ks['pledged'] > 0

df_ks = df_ks[cond1]

df_ks.head(10)
```

### 5. Collect only those projects that were hosted in the US

```python
cond1 = df_ks['country'] == 'US'

df_ks = df_ks[cond1]

df_ks.head(10)
```

### 6. Create a new column that finds the average amount pledged to a project

```python
df_ks['average_donation'] = df_ks['pledged'] / df_ks['backers_count']

df_ks.head(5)
```

### 7. First convert "average_donation", "goal", and "pledged" columns to float Then Format to go to two decimal places, include a dollar sign, and use comma notation

```python
df_ks.info()
```

```python
df_ks['average_donation'] = df_ks['average_donation'].map("${:,.2f}".format)
df_ks['goal'] = df_ks['goal'].map("${:,.2f}".format)
df_ks['pledged'] = df_ks['pledged'].map("${:,.2f}".format)

df_ks.head(5)
```

### 8. Calculate the total number of backers for all US projects 

```python
total_backers = df_ks['backers_count'].sum()

total_backers
```

### 9. Calculate the average number of backers for all US projects

```python
avg_backers = df_ks['backers_count'].mean()

avg_backers
```

### 10. Collect only those US campaigns that have been picked as a "Staff Pick"

```python
cond1 = df_ks['staff_pick'] == True

df_ks = df_ks[cond1]

df_ks.head(10)
```

### 11. Group by the state of the campaigns and see if staff picks matter (Seems to matter quite a bit)

```python
df_ks.groupby(by='state').count()
```
